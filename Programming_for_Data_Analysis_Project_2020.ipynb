## First Impressions 

The old saying, "First impressions are everything" rings true in many cases. Sadly, many interviewers will decide to not hire a candidate within the first 30 seconds to 3 minutes, based on a poor first impression; so don't hurt your chances by arriving at an interview and making a poor first impression. Certain interviewers do put far too much emphasis on first impressions, this is the "human factor" which is hard to factor into the success rate, however this is the reality of the hiring process [10].



McIntosh (2014), in his book titled  "Job Interview Success for Introverts" identified the following key variables as having a major impact on first impressions, and subsequent to this a dramatic effect on the chances of success for the applicant. In order to consider first impressions eye contact, the handshake, the smile and the posture of applicants will be simulated [10]:

- **Eye Contact.** Making steady eye contact is essential at interviews, as it tells the interviewer(s) that you are trustworthy, engaged, interested, enthusiastic, and aren't hiding something. Your eye contact is part of your communications; in other words, you can speak with your eyes. Eyes wide open can indicate excitement and happiness, while slightly closed eyes can indicate disapproval and even anger


- **The Handshake.** This is quite topical now and certainly given the unprecedented situation with COVID,now we see  the elements concerning first impression once centerted on the handshake are now negated. However for the sake of this assignment we will continue as normal. The handshake is just as important as eye contact, its a form of communication. The general rule on handshakes is to deliver one that is firm yet gentle. A hand shake that is too firm can be seen as aggressive, while a handshake that is to gentle can be seen as weak or cunning. 


- **The Smile.** A smile tells the interviewers that you're happy to be there. Most people are naturally drawn more to people who smile than those who don't. But don't force a smile that comes across as fake.


- **Your Posture.** How you sit is also important. Interviewers will observe your posture and body language, so be cognizant of how straight you're sitting and if you are fidgeting with your fingers or hands or even tapping your feet


- **Small Talk.** Your ability to engage in small talk speaks to your verbal communication skills, ability to reason, and how well you think quickly on your feet. If this is important to the interviewer(s), they'll be paying close attention to how well you can interact quickly and naturally
