```python
# Call the fake name library 
fake = Faker()
fake.name()

print(fake.name())

# Create an empty list
Fake_Name_List = []
# Create a for loop 1500 times
for i in range(1500):
    # Append fake naems to list 
    Fake_Name_List.append(fake.name())
    # Print names
    print(fake.name())
```

```python
# Create DataFrame to store the fake names
df_Names = pd.DataFrame(Fake_Name_List, columns=["Names"])
df_Names
```

```python
# Generate a random normal distribution, centered on 4hrs (approx 60% of entire data), 3 std,/ 
# and size of 1500 data points as per ref[2].
Hrs_prep = np.random.normal(4, 3, 1500)
Hrs_prep
```

```python
# Create DataFrame to store the Hours Preparation based on a normal distribution
df = pd.DataFrame(Hrs_prep, columns=["Hours_Preparation"])
df
```

Normal Distributions can be used to represent random distribution for a wide range of real-world applications. Normal distributions have the following features:

-symmetric bell shape [3]:

-mean and median are equal; both located at the center of the distribution

-Approx 68% of the data falls within 1 standard deviation of the mean

-Approx 95% of the data falls within 2 standard deviations of the mean

-Approx 99.7% of the data falls within 3 standard deviations of the mean

The internet has made it very easy for people to search out and apply to many job opportunities. But sending out more applications doesn’t increase your chances of getting hired.
```python
# Controlling the graph so that it flows in descending volume [6]
sns.countplot(x="Qualifications", data=df_1, palette="Set3", order = df_1["Qualifications"].value_counts().index)
plt.show()
```

From the above depictions and graphs we can see that the distribution is clearly descending. 25% of applicants are classed as being qualified, while 75% are only somewhat qualified, are under qualified of have no qualifications at all. Therefore we would expect to see the majority of volume contained in these columns. ## Creating a Machine Learning Algorithm to Predict the Success of Applicants based on First Impressions

I want to use K nearest neighbors to try and predict the success (pass/fail) of applicants based on their variable of first impressions. I will use the 5 identified variables to try and predict the pass or fail. 

```python
# import K nearest neighbors
import sklearn.neighbors as nei
```

### Inputs and Outputs

```python
# seperate the data into inputs and outputs

# inputs
inputs =df_First_Impressions[["Good_Eye_Contact","Good_Handshake", "Good_Smile", "Good_Posture", "Good_Talk"]]

# convert True/False Bools to Integers 
inputs.astype(int)
```

```python
# outputs 
outputs = df_First_Impressions[["Pass_Fail_First_Impressions"]]
outputs
```

### Classifier 

```python
# create classifier that uses 5 points of data to make prediction, store as "knn"
knn = nei.KNeighborsClassifier(n_neighbors=5)
```

### Fit

```python
# train classifier on inputs & outputs data
# .values will give the values in an array (shape: (n,1)
# .ravel will convert that array shape to (n, )
knn.fit(inputs, outputs.values.ravel())
```

### Predict 

```python
# Call up first row of data (4x True, 1x False = Fail. 5x True = Pass)
df_First_Impressions.loc[0]
```

```python
# Call up fourth row of data  (5x true = pass)
df_First_Impressions.loc[3]
```

```python
# use Knn with 2 sets of inputs. One set should return a fail and one should return a pass (if KNN is accurate)
# we can see that it has predicted a fail which is correct
knn.predict([[1,1,1,1,0],[1,1,1,1,1]])
```

### Evaluate 

```python
# import model_selection from sklearn, store as mod
import sklearn.model_selection as mod
```

```python
# take the entire DataFrame 1500 rows, split into inputs & outputs
# split it into 2 parts- inputs_train & outputs_train
# puts rest of set into inputs_test & outputs_test- sample size 33% (500)
inputs_train, inputs_test, outputs_train, outputs_test = mod.train_test_split(inputs, outputs, test_size=0.33)
```

```python
# we can see 1005 rows in which we haved trained inputs on- inputs_train
inputs_train
```

```python
# We also have the other 1005 rows as outputs_train
outputs_train
```

```python
# train classifier on the inputs_train data
knn = nei.KNeighborsClassifier(n_neighbors=5)
knn.fit(inputs_train, outputs_train.values.ravel())
```

```python
# test on classifier on the inputs_test 
knn.predict(inputs_test) 
```

```python
outputs_test 
```

```python
# We can now use the Knn to make predictions based on the inputs data
# Knn has been rained on the input_test data
np.set_printoptions(threshold=np.inf)
knn.predict(inputs)
```

```python
# Count pass and fail rate totals from the classifier
Counter(knn.predict(inputs))
```

```python
# Count pass and fail rate totals from the actual data column
Counter(df_First_Impressions["Pass_Fail_First_Impressions"])
```

The algorithm has returned an exact match of the actual pass and fail rate with 100% accuracy.
