## Average Hours Spent in Preparation for a Job Interview

According to the Global Career Hub and the Association of International Certified Accountants, about 60% of persons spend four hours or less preparing for a job interview. About 28% spend between five and nine hours preparing for a job interview. Only 12% spend ten hours or more preparing for a job interview. This data was complied from 1,500 survey answers [2].  Its interesting that its suggested a minimum of ten hours is required to prepare for an interview, yet by their own research only 12% of persons are willing to put in such effort. This answer correlates exactly with the Google search "how many hours to spend in preparation for interview"



Although this is by no means a definitive answer it does provide a good approximation of the preparation required that is generally recommended. 

```python
# Generate a random normal distribution, centered on 4hrs (approx 60% of entire data), 3 std,/ 
# and size of 1500 data points as per ref[2].
Hrs_prep = np.random.normal(4, 3, 1500)
Hrs_prep
```

```python
df = pd.DataFrame(Hrs_prep, columns=["Hours_Preparation"])
df
```

```python
# Visualise normal distribution, data used form Hrs_prep  random data, centred at 4
sns.histplot(data=Hrs_prep, kde=True)

# Generate a distplot with labels to visualise.
plt.title("Number of Hours Spent in Preparation")
plt.xlabel("Hours")
plt.ylabel("Number of Persons")
plt.show()
```

We can see here that the Hours spent in preparation for an interview follows a Normal Distribution curve. 

Normal Distributions can be used to represent random distribution for a wide range of real-world applications. Normal distributions have the following features:

-symmetric bell shape [3]:

-mean and median are equal; both located at the center of the distribution

-Approx 68% of the data falls within 1 standard deviation of the mean

-Approx 95% of the data falls within 2 standard deviations of the mean

-Approx 99.7% of the data falls within 3 standard deviations of the mean

```python
# At this point I realise that I can not have "negative study hours" and I possibly made a mistake when selecting the distribution.
# I will try a Log Noraml as this is seems to be a better fit to the data ref [2].

# Generate Log Normal Distribution 
Hrs_prep_1 = np.random.lognormal(1.5, .4, 1500)
Hrs_prep_1
```

```python
# Visualise normal distribution, data used form Hrs_prep  random data, centred at 4
sns.histplot(data=Hrs_prep_1, kde=True)

# Generate a distplot with labels to visualise.
plt.title("Number of Hours Spent in Preparation")
plt.xlabel("Hours")
plt.ylabel("Number of Persons")
plt.show()
```

Some descriptive figures- further analysis. Mean, Median, Std Dev, Min & Max Hrs.

```python
# Mean Hrs study - this is extremely close to the 4hour or less study time on average/
# that people set aside for interview preparation. Therefore the Log Normal Distribution is acurrate fit
np.mean(Hrs_prep_1)
```

```python
# Median Hrs study
np.median(Hrs_prep_1)
```

```python
# Standard Dev
np.std(Hrs_prep_1)
```

```python
# Average Min Hrs study 
np.min(Hrs_prep_1)
```

```python
# Average Max Hrs study
np.max(Hrs_prep_1)
```

A vast number of employers have told Workopolis Work that as many as 75% of job applicants for a given role aren’t actually qualified to do it! Furthermore, they state that an even smaller fraction than that are selected for an interview [5].

“98% of job seekers are eliminated at the initial resume screening and only the Top 2% of candidates make it to the interview”, says Robert Meier, President of Job Market Experts. “Fixing the employment market requires helping job seekers become “Top 2% Candidates” who can meet employer’s rigorous requirements and easily hit the “bulls-eye” of employer needs to ensure they don’t make bad hires” continued Meier [5].

```python
# Create  a list of Qualifications 
Qualifications = ["None", "Under_Qualified", "Somewhat_Qualified", "Qualified", "Over_Qualified"]
Qualifications
```
